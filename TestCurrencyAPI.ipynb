```python
# These are POC tests to get the exchange rate for USD to ZAR using the ExchangeRate-API.

import requests

EXCHANGE_URL = "https://v6.exchangerate-api.com/v6/529d3453809b9ec1e660421b/latest/USD"

response = requests.get(EXCHANGE_URL)

exchange_data = response.json()
exchange_rate_for_USD_to_rand = exchange_data['conversion_rates']['ZAR']

print(exchange_rate_for_USD_to_rand)
```
