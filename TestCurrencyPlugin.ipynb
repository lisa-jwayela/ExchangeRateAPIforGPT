```python
# These are tests to call the API created in this project and hosted on pythonanywhere.com 
# to get the exchange rate for USD to ZAR using the ExchangeRate-API.

import requests

# Configuration
SERVICE_AUTH_KEY = "3987598723758730397456"
API_BASE_URL = "https://lisajwa.pythonanywhere.com/"  
# Set up headers with Bearer token
headers = {
    "Authorization": f"Bearer {SERVICE_AUTH_KEY}"
}

# Call the USD Rate endpoint
response = requests.get(f"{API_BASE_URL}/USDRate", headers=headers)

returnstring = ""
# Check if request was successful
if response.status_code == 200:
    usd_to_zar_rate = response.text
    returnstring = f"USD to ZAR Exchange Rate: {usd_to_zar_rate}"
else:
    returnstring = f"Error: {response.text}"

print(returnstring)
```
