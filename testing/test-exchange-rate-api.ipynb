```python
# Integration sanity check: call ExchangeRate-API directly for USD -> ZAR.

import os
import requests

api_key = os.getenv("EXCHANGE_RATE_API_KEY")
if not api_key:
    raise ValueError(
        "Set EXCHANGE_RATE_API_KEY with Command Prompt before running this notebook cell."
    )

EXCHANGE_URL = f"https://v6.exchangerate-api.com/v6/{api_key}/latest/USD"

response = requests.get(EXCHANGE_URL)

exchange_data = response.json()
exchange_rate_for_USD_to_rand = exchange_data["conversion_rates"]["ZAR"]

print(exchange_rate_for_USD_to_rand)
```
