```python
# API wrapper test: call this project endpoint for GBP -> ZAR on PythonAnywhere.

import os
import requests

# Configuration
service_auth_key = os.getenv("SERVICE_AUTH_KEY")
if not service_auth_key:
    print(service_auth_key)
    raise ValueError("Set SERVICE_AUTH_KEY before running this notebook cell.")

API_BASE_URL = "https://lisajwa.pythonanywhere.com"
headers = {"Authorization": f"Bearer {service_auth_key}"}

# Call the GBP Rate endpoint
response = requests.get(f"{API_BASE_URL}/GBPRate", headers=headers)

# Check if request was successful
if response.status_code == 200:
    gbp_to_zar_rate = response.text
    print(f"GBP to ZAR Exchange Rate: {gbp_to_zar_rate}")
else:
    print(f"Error ({response.status_code}): {response.text}")
```

```python
# API wrapper test: call this project endpoint for USD -> ZAR on PythonAnywhere.

import os
import requests

# Configuration
service_auth_key = os.getenv("SERVICE_AUTH_KEY")
if not service_auth_key:
    raise ValueError("Set SERVICE_AUTH_KEY before running this notebook cell.")

API_BASE_URL = "https://lisajwa.pythonanywhere.com"
headers = {"Authorization": f"Bearer {service_auth_key}"}

# Call the USD Rate endpoint
response = requests.get(f"{API_BASE_URL}/USDRate", headers=headers)

if response.status_code == 200:
    usd_to_zar_rate = response.text
    return_string = f"USD to ZAR Exchange Rate: {usd_to_zar_rate}"
else:
    return_string = f"Error ({response.status_code}): {response.text}"

print(return_string)
```
